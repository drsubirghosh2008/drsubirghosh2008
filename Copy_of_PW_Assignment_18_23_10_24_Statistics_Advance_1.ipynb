Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

Answer:



```python
import math

def normal_pdf(x, mu, sigma):
    """
    Calculate the probability density function of a normal distribution.

    Parameters:
    x (float): The point at which to evaluate the PDF.
    mu (float): The mean of the distribution.
    sigma (float): The standard deviation of the distribution.

    Returns:
    float: The value of the PDF at the given point x.
    """
    if sigma <= 0:
        raise ValueError("Standard deviation must be positive.")

    coefficient = 1 / (math.sqrt(2 * math.pi) * sigma)
    exponent = -((x - mu) ** 2) / (2 * sigma ** 2)

    return coefficient * math.exp(exponent)

# Example usage:
mean = 0
std_dev = 1
point = 0
pdf_value = normal_pdf(point, mean, std_dev)
print(f"The PDF value at x={point} for N({mean}, {std_dev}) is {pdf_value:.4f}")
```

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

Answer:

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

Answer:

To calculate the cumulative distribution function (CDF) of a Poisson distribution using the following Python function. The CDF gives the probability that a Poisson-distributed random variable is less than or equal to a specific value x.

Here's how you can implement it using the scipy.stats library, which provides a convenient way to work with probability distributions:

```python
import scipy.stats as stats

def poisson_cdf(lam, x):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
    lam (float): The mean (lambda) of the Poisson distribution.
    x (int): The point at which to calculate the CDF.

    Returns:
    float: The CDF of the Poisson distribution at the point x.
    """
    if x < 0:
        return 0.0  # CDF is 0 for negative x
    return stats.poisson.cdf(x, lam)

# Example usage:
mean = 3.5  # example mean
point = 5   # example point
cdf_value = poisson_cdf(mean, point)
print(f"The CDF of the Poisson distribution at x={point} with mean={mean} is: {cdf_value:.4f}")
```

lam is the mean of the Poisson distribution (often denoted as λ).
x is the point at which you want to evaluate the CDF.
The function uses stats.poisson.cdf from the scipy.stats library to compute the CDF efficiently.

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

n a normal distribution, the least frequent data points appear at the extremes, far from the mean. Here’s a breakdown:

Characteristics of Normal Distribution

Symmetrical Bell Curve: The normal distribution is symmetrical around its mean, meaning the data is evenly distributed on either side of the mean.

Mean, Median, and Mode: In a normal distribution, the mean, median, and mode are all located at the center of the distribution.

Position of Least Frequent Data

Extremes: The least frequent data points occur in the tails of the distribution, which are the areas farthest from the mean. Specifically:

The left tail (values much less than the mean)

The right tail (values much greater than the mean)

Probability Density: As you move away from the mean towards either tail, the probability density decreases. This means that data points far from the mean are less likely to occur.


In a normal distribution, the least frequent data points are located at the extremes (in the left and right tails), far from the mean, where the probability of occurrence is lowest.
