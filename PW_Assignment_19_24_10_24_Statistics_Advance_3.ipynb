Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Answer:

Estimation statistics is a data analysis framework that uses confidence intervals, effect sizes, and other methods to plan experiments, analyze data, and interpret results.

Point estimation and interval estimation are two types of estimates used in estimation statistics:

Point estimation

Uses a single value, or sample statistic, to estimate an unknown population parameter. For example, the average age of a sample is a point estimate of the population mean.

Interval estimation

Uses a range of numbers to estimate the range of values where a population parameter is likely to lie. The most common type of interval estimate is a confidence interval.

Here are some more details about point and interval estimation:

Accuracy

Point estimates are specific values, so the accuracy of the estimate cannot be determined.

Confidence level

The confidence level of an interval estimate is denoted by 1 - alpha, where alpha is a value between 0 and 1. For example, if you want to be 95% confident that the parameter is inside the interval, alpha is 5%.

Margin of error

The margin of error is a component of a confidence interval. It's calculated by adding the standard error of the sample statistic to a multiplier based on the desired level of confidence

Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

Answer:



```python
import math
from scipy import stats

def estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level=0.95):
    # Calculate the standard error of the mean
    standard_error = sample_std_dev / math.sqrt(sample_size)

    # Calculate the critical value (z or t) depending on the sample size
    if sample_size >= 30:
        # Use Z-distribution (normal distribution) for large samples
        critical_value = stats.norm.ppf((1 + confidence_level) / 2)
    else:
        # Use T-distribution for small samples
        critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size-1)

    # Calculate margin of error
    margin_of_error = critical_value * standard_error

    # Estimate population mean (with confidence interval)
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_mean = 50
sample_std_dev = 10
sample_size = 25
confidence_level = 0.95

lower_bound, upper_bound = estimate_population_mean(sample_mean, sample_std_dev, sample_size, confidence_level)

print(f"Estimated population mean is between {lower_bound:.2f} and {upper_bound:.2f}")
```

Explanation:

Sample Mean: This is the average of the sample data points.

Sample Standard Deviation: This measures the spread of the sample data points.

Standard Error of the Mean (SEM): This is the sample standard deviation divided by the square root of the sample size. It quantifies how much the sample mean deviates from the population mean.

Confidence Interval: We calculate a confidence interval to estimate the range within which the true population mean lies. For small sample sizes, we use the t-distribution, and for large samples, we use the z-distribution.

The function returns the lower and upper bounds of the estimated population mean.

Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

Answer:

Hypothesis Testing:

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating two competing hypotheses, a null hypothesis (H‚ÇÄ) and an alternative hypothesis (H‚ÇÅ), and then using sample data to determine which hypothesis is more supported by the evidence.

Null Hypothesis (H‚ÇÄ): A statement that there is no effect or no difference, often representing the status quo or default assumption.

Alternative Hypothesis (H‚ÇÅ): A statement that contradicts the null hypothesis, suggesting that there is an effect, a difference, or an association.

Why is Hypothesis Testing Used?

Hypothesis testing is used to determine if there is enough evidence in a sample to infer that a certain condition or relationship exists in the population. It helps in:

Decision Making: It aids in deciding whether to accept or reject the null hypothesis based on sample data.

Testing Claims: Hypothesis testing allows researchers to test claims or theories about population parameters (e.g., population mean, proportion).
Statistical Inference: It enables generalizing findings from a sample to the larger population.

Steps in Hypothesis Testing:

State the Hypotheses: Formulate the null and alternative hypotheses.
Choose the Significance Level (Œ±): Decide on the significance level, typically 0.05 or 5%.

Collect Sample Data: Gather data from a sample relevant to the hypothesis.
Calculate the Test Statistic: Compute the test statistic (e.g., z-score, t-score) based on the sample data.

Make a Decision: Compare the test statistic with critical values or p-values to decide whether to reject or fail to reject the null hypothesis.

Importance of Hypothesis Testing:

Evidence-Based Decision Making: It provides a structured framework for making decisions based on empirical evidence, reducing the influence of bias or intuition.

Validation of Assumptions: Hypothesis testing validates or refutes assumptions about a population parameter based on sample data, ensuring more accurate conclusions.

Quantification of Uncertainty: Hypothesis testing quantifies the likelihood of making errors, particularly Type I error (false positive) and Type II error (false negative), thus providing a clearer picture of the risks involved in decision-making.

Scientific Research: In research, hypothesis testing is a cornerstone for testing new theories, treatments, or interventions. It allows scientists to confirm or disprove hypotheses based on experimental or observational data.

Quality Control: In industries, hypothesis testing is often used for quality assurance, allowing businesses to make informed decisions about products or processes based on sample inspections.

Example:

In medical research, hypothesis testing could be used to determine if a new drug is more effective than an existing one. The null hypothesis (H‚ÇÄ) might state that there is no difference in effectiveness between the two drugs, while the alternative hypothesis (H‚ÇÅ) would suggest that the new drug is more effective.

In summary, hypothesis testing is a critical tool in statistics, research, and various industries to make informed, data-driven decisions while controlling the risk of incorrect conclusions.

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.

Answer:


To conduct a hypothesis test on the difference between two population means using a sample from each population, you can use a two-sample t-test. This test is used when the population standard deviations are unknown, and you're testing if there is a significant difference between the means of two populations.


```python
import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    """
    Conducts a two-sample t-test to compare the means of two populations.

    Args:
    sample1 (array-like): The first sample (e.g., from population 1)
    sample2 (array-like): The second sample (e.g., from population 2)
    alpha (float): The significance level (default is 0.05)

    Returns:
    t_statistic (float): The computed t-statistic
    p_value (float): The p-value of the test
    conclusion (str): Conclusion on whether to reject the null hypothesis
    """
    # Calculate the sample means and standard deviations
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)

    # Perform a two-sample t-test (assumes unequal variances by default)
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    # Print the results
    print(f"Sample 1 mean: {mean1:.2f}")
    print(f"Sample 2 mean: {mean2:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Compare the p-value to the significance level (alpha) to make a decision
    if p_value < alpha:
        conclusion = "Reject the null hypothesis (H‚ÇÄ): There is a significant difference between the two means."
    else:
        conclusion = "Fail to reject the null hypothesis (H‚ÇÄ): No significant difference between the two means."

    return t_statistic, p_value, conclusion

# Example usage:
# Sample data from two populations
sample1 = [85, 88, 90, 93, 86, 91, 89]
sample2 = [82, 83, 85, 87, 84, 88, 86]

# Perform the t-test
t_stat, p_val, result = two_sample_t_test(sample1, sample2)

# Output the conclusion
print(result)
```

Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

Answer:

Below is a Python program that calculates the two-sample t-test for independent samples, assuming the null hypothesis that the population means are equal.

```python
import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2):
    """
    Conducts a two-sample t-test for independent samples.

    Args:
    sample1 (array-like): First sample data
    sample2 (array-like): Second sample data

    Returns:
    t_statistic (float): The computed t-statistic
    p_value (float): The p-value of the test
    conclusion (str): The decision on the null hypothesis
    """
    # Perform the t-test for independent samples
    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    # Print the results
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Set significance level (alpha)
    alpha = 0.05

    # Make a decision based on the p-value
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: There is a significant difference between the two population means."
    else:
        conclusion = "Fail to reject the null hypothesis: No significant difference between the two population means."

    return t_statistic, p_value, conclusion

# Example usage:
# Generate two random samples of equal size
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=5, size=30)  # Sample from population 1 (mean=50, std=5)
sample2 = np.random.normal(loc=52, scale=5, size=30)  # Sample from population 2 (mean=52, std=5)

# Conduct the two-sample t-test
t_stat, p_val, result = two_sample_t_test(sample1, sample2)

# Output the conclusion
print(result)
```

Explanation:

Function two_sample_t_test:

Takes two independent samples (sample1 and sample2).
Uses scipy.stats.ttest_ind to calculate the t-statistic and p-value, assuming unequal variances (equal_var=False).
Compares the p-value to the significance level (default is 0.05).

Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.

Estimate the population mean revenue with a 95% confidence interval.

Answer:

To estimate the population mean revenue with a 95% confidence interval using the sample data, we can follow these steps:

### Step 1: Identify the sample statistics
- Sample mean (\( \bar{x} \)) = $500
- Sample standard deviation (\( s \)) = $50
- Sample size (\( n \)) = 50

### Step 2: Determine the critical value
Since the sample size is large (\( n \geq 30 \)), we can use the z-distribution. For a 95% confidence level, the critical value (z*) is approximately 1.96.

### Step 3: Calculate the standard error (SE)
The standard error (SE) is calculated using the formula:

\[
SE = \frac{s}{\sqrt{n}} = \frac{50}{\sqrt{50}} \approx 7.07
\]

### Step 4: Calculate the margin of error (ME)
The margin of error (ME) can be calculated as follows:

\[
ME = z^* \times SE = 1.96 \times 7.07 \approx 13.87
\]

### Step 5: Calculate the confidence interval
The confidence interval can be computed using the sample mean and the margin of error:

\[
\text{Confidence Interval} = \bar{x} \pm ME = 500 \pm 13.87
\]

### Result
Thus, the 95% confidence interval for the population mean revenue is:

\[
(500 - 13.87, 500 + 13.87) \approx (486.13, 513.87)
\]

Therefore, the 95% confidence interval for the average daily revenue is approximately **($486.13, $513.87)**.
