Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

Answer:

In statistics, a Type I error means rejecting the null hypothesis when it's actually true, while a Type II error means failing to reject the null hypothesis when it's actually false.

In simple terms, a type I error is a false positive result. If a person was diagnosed with a medical condition that they do not have, this would be an example of a type I error. Similarly, if a person was convicted of a crime, a type I error occurs if they were innocent.

A type II error produces a false negative, also known as an error of omission. For example, a test for a disease may report a negative result when the patient is infected. This is a type II error because we accept the conclusion of the test as negative, even though it is incorrect.

Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.

Answer:

To calculate the 90% confidence interval for the true population mean, we can use the formula:

\[
CI = \bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
\]

Where:
- \(\bar{x} = 65\) is the sample mean
- \(\sigma = 8\) is the population standard deviation
- \(n = 50\) is the sample size
- \(Z_{\alpha/2}\) is the critical value for a 90% confidence level, which is 1.645

### Step-by-Step Calculation:

1. **Identify the values**:
   - Sample mean (\(\bar{x}\)) = 65
   - Population standard deviation (\(\sigma\)) = 8
   - Sample size (\(n\)) = 50
   - Critical value (\(Z_{\alpha/2}\)) = 1.645 for a 90% confidence level

2. **Calculate the standard error (SE)**:
   \[
   SE = \frac{\sigma}{\sqrt{n}} = \frac{8}{\sqrt{50}} = \frac{8}{7.071} \approx 1.131
   \]

3. **Calculate the margin of error (MOE)**:
   \[
   MOE = Z_{\alpha/2} \cdot SE = 1.645 \times 1.131 \approx 1.860
   \]

4. **Calculate the confidence interval**:
   \[
   CI = \bar{x} \pm MOE = 65 \pm 1.860
   \]
   \[
   CI = (65 - 1.860, 65 + 1.860)
   \]
   \[
   CI = (63.14, 66.86)
   \]

### Interpretation:

The 90% confidence interval for the true population mean is **(63.14, 66.86)**. This means we are 90% confident that the true population mean lies within this interval, based on the sample data.

Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

Answer:

To determine if caffeine has a significant effect on reaction time, we can perform a **one-sample t-test**. This test will assess whether the mean reaction time of the sample is significantly different from a hypothesized population mean (which we assume to be the typical reaction time without caffeine).

### Step-by-Step Approach

#### 1. **State the Hypotheses:**

- **Null hypothesis (\(H_0\))**: Caffeine has no effect on reaction time, so the mean reaction time is equal to the hypothesized population mean (\(\mu_0\)).
  - \(\mu = \mu_0\)
- **Alternative hypothesis (\(H_1\))**: Caffeine has a significant effect on reaction time, so the mean reaction time is different from the population mean (\(\mu_0\)).
  - \(\mu \neq \mu_0\)

If no value for the population mean is provided, we assume that without caffeine, the average reaction time would be a known baseline, say \( \mu_0 = 0.30 \) seconds (you can adjust this value if needed). This is a **two-tailed test** because we want to test whether caffeine affects reaction time either positively or negatively.

#### 2. **Set the Significance Level:**

- The confidence level is 90%, so the significance level (\(\alpha\)) is 0.10.

#### 3. **Calculate the t-statistic:**

The formula for the t-statistic is:

\[
t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}}
\]

Where:
- \(\bar{x} = 0.25\) is the sample mean
- \(\mu_0 = 0.30\) is the hypothesized population mean (without caffeine)
- \(s = 0.05\) is the sample standard deviation
- \(n = 30\) is the sample size

Let's plug in the values:

\[
t = \frac{0.25 - 0.30}{\frac{0.05}{\sqrt{30}}} = \frac{-0.05}{\frac{0.05}{5.477}} = \frac{-0.05}{0.00913} \approx -5.48
\]

#### 4. **Determine the Critical t-value:**

- Degrees of freedom \(df = n - 1 = 30 - 1 = 29\).
- For a two-tailed test at \(\alpha = 0.10\) with \(df = 29\), the critical t-value from a t-distribution table is approximately **Â±1.699**.

#### 5. **Make a Decision:**

- If the calculated t-statistic is greater than the critical t-value (in magnitude), we reject the null hypothesis.
- In this case, the calculated \(t = -5.48\) is much larger in magnitude than the critical t-value of 1.699.

#### 6. **Conclusion:**

Since the calculated t-statistic exceeds the critical value, we **reject the null hypothesis**. This means there is significant evidence at the 90% confidence level to suggest that caffeine has a significant effect on reaction time.

### Interpretation:

The result of the hypothesis test indicates that caffeine significantly affects reaction time, as the sample mean reaction time of 0.25 seconds is significantly different from the assumed population mean of 0.30 seconds at the 90% confidence level.
