Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Answer:

Here are five common plots that can be created using the Seaborn library, along with their uses:

Line Plot:

Use: Displays trends or changes over time by connecting data points with lines. Line plots are ideal for visualizing data with a sequential order, like time series.

Bar Plot:

Use: Shows comparisons between discrete categories. Itâ€™s useful for comparing quantities or counts across different groups, such as sales by category or population by region.

Scatter Plot:

Use: Visualizes the relationship between two continuous variables, often highlighting correlations or clusters. Scatter plots are widely used in regression analysis and exploratory data analysis.

Histogram:

Use: Displays the distribution of a single variable by grouping data into bins. Histograms are useful for understanding the underlying frequency distribution of the data, such as the age distribution in a population.

Heatmap:

Use: Represents data as a matrix of colors, typically used to show correlation matrices, feature relationships, or data densities. Heatmaps are excellent for visualizing complex relationships in high-dimensional data, especially in machine learning or correlation analyses.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

Answer:



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
# Create a line plot with 'timepoint' on x-axis, 'signal' on y-axis, and separate lines for each 'event' and 'region'
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event and Region")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.


Note: pclass, age, and fare are columns in the titanic dataset.

Answer:



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
try:
    # Set up the matplotlib figure with subplots
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Plot box plot for age by pclass
    sns.boxplot(data=titanic, x='pclass', y='age', ax=axes[0])
    axes[0].set_title('Age Distribution by Passenger Class')
    axes[0].set_xlabel('Passenger Class')
    axes[0].set_ylabel('Age')

    # Plot box plot for fare by pclass
    sns.boxplot(data=titanic, x='pclass', y='fare', ax=axes[1])
    axes[1].set_title('Fare Distribution by Passenger Class')
    axes[1].set_xlabel('Passenger Class')
    axes[1].set_ylabel('Fare')

    plt.tight_layout()
    plt.show()

except Exception as e:
    print(e) # Print the exception for debugging
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Answer:



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")
flights.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot the data to create a matrix format for the heatmap
flights_pivot = flights.pivot(index = "month", columns = "year", values = "passengers")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d", cbar_kws={'label': 'Number of Passengers'})
plt.title("Monthly Number of Passengers (1949 - 1960)")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
