Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Answer:


Matplotlib is a popular Python library used for data visualization. It provides tools for creating static, animated, and interactive plots in Python. Its main module, pyplot, simplifies the process of generating various types of plots by providing a MATLAB-like interface, making it easier to generate complex visualizations with minimal code.

Why is Matplotlib used?


Matplotlib is used to:

1. Visualize data to understand patterns, trends, and outliers.

2. Represent complex datasets in a clear and interpretable way.

3. Customize plots extensively to make them publication-ready.

4.Facilitate interactive data exploration in Jupyter Notebooks or interactive environments.

Five Plots That Can Be Plotted Using pyplot:

Line Plot - Used to display data as a series of points connected by straight lines.

Bar Plot - Represents categorical data with rectangular bars.

Histogram - Shows the distribution of a dataset by grouping data into bins.

Scatter Plot - Displays individual data points on a two-dimensional plane, often used to observe relationships.

Pie Chart - Represents proportions of a whole, showing parts of a dataset as slices of a circle.

These are just a few examples; Matplotlib supports many other plot types for various data visualization needs.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Answer:



The scatter ia a plot of data that shows randomly distributed points in two dimensions, illustrating the spread and relationship between the x and y values

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting a scatter plot
plt.scatter(x, y, color="blue", alpha=0.7)
plt.title("Scatter Plot of Randomly Generated Data")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer:

The subplot() function is used to create multiple plots in a single figure, arranging them in a grid layout. This allows us to visualize different datasets or perspectives side by side, which is particularly useful for comparison.

```python
# Importing required library
import numpy as np
import matplotlib.pyplot as plt

# Data for each line plot
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots for each line
plt.figure(figsize=(10, 8))

# Line plot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, color="blue", marker="o")
plt.title("Line Plot 1")
plt.xlabel("X values")
plt.ylabel("Y values")

# Line plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, color="red", marker="o")
plt.title("Line Plot 2")
plt.xlabel("X values")
plt.ylabel("Y values")

# Line plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, color="green", marker="o")
plt.title("Line Plot 3")
plt.xlabel("X values")
plt.ylabel("Y values")

# Line plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, color="purple", marker="o")
plt.title("Line Plot 4")
plt.xlabel("X values")
plt.ylabel("Y values")

# Displaying all subplots
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Answer:

A bar plot is a type of chart used to represent categorical data with rectangular bars. Each bar’s height (or length, in the case of horizontal bars) represents the value of that category, making it easy to compare different groups or categories. Bar plots are commonly used for comparing discrete data points or summarizing values across different categories.

Here, both a vertical and a horizontal bar plot display company profits, with each company represented by a bar proportional to its profit value

```python
# Importing required library
import numpy as np
import matplotlib.pyplot as plt

# Data for the bar plots
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating the bar plot
plt.figure(figsize=(10, 5))

# Vertical bar plot
plt.subplot(1, 2, 1)
plt.bar(company, profit, color="blue")
plt.title("Company Profits - Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit ($)")

# Horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color="green")
plt.title("Company Profits - Horizontal Bar Plot")
plt.xlabel("Profit ($)")
plt.ylabel("Company")

# Displaying the plots
plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot. box1 = np.random.normal(100, 10, 200) box2 = np.random.normal(90, 20, 200)

Answer:

A box plot (or box-and-whisker plot) is a statistical chart that visually represents the distribution of a dataset through its quartiles. It displays key summary statistics: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values. Outliers are often marked as individual points.

Box plots are useful for:

Comparing distributions between different groups.

Identifying data spread, central tendency, and any outliers.


```python
# Data for box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"], patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Box Plot of Two Distributions")
plt.xlabel("Box Category")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
