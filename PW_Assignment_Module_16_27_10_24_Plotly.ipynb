Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

Answer:



```python
import seaborn as sns
import plotly.express as px

# Load the "titanic" dataset from seaborn
titanic_data = sns.load_dataset("titanic")
titanic_data.head()
```

```python
# Plotting a scatter plot for 'age' vs 'fare' using Plotly Express
fig = px.scatter(titanic_data, x="age", y="fare", title="Scatter Plot of Age vs Fare",
                 labels={"age": "Age", "fare": "Fare"}, opacity=0.6)
fig.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

Answer:



```python
import plotly.express as px

# Load the "tips" dataset directly from Plotly's library
tips_data = px.data.tips()
tips_data.head()
```

```python
# Creating a box plot using Plotly Express for 'total_bill' grouped by 'day'
fig = px.box(tips_data, x="day", y="total_bill", color="day", title="Box Plot of Total Bill by Day",
             labels={"day": "Day of the Week", "total_bill": "Total Bill"})
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

Answer:



```python
import plotly.express as px
import plotly.data as data

# Load the tips dataset
tips = data.tips()

# Create the histogram
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    title="Total Bill by Sex with Day and Smoker Pattern",
    labels={"total_bill": "Total Bill", "sex": "Sex"}
)

# Show the plot
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.

Answer:



```python
# Importing necessary libraries
import plotly.express as px

# Load the iris dataset
iris = px.data.iris()
iris.head()
```

```python
# Importing necessary libraries
import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Create the scatter matrix plot
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix of Iris Dataset by Species"
)

# Show the plot
fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Answer:

In data visualization, a Distplot (or distribution plot) is used to display the distribution of a dataset, usually by showing a histogram combined with a density curve. This type of plot helps to visualize the shape, spread, and central tendency of the data, providing insight into its underlying distribution.

While Plotly Express does not directly offer a distplot function, it can achieve similar results by using px.histogram along with the marginal parameter set to "rug" or "violin" to provide additional distribution detail. For a classic distplot look, the histogram can be combined with density information from Seaborn.

```python
# example of how to create a distribution plot in Plotly Express:
import plotly.express as px
import plotly.figure_factory as ff
import numpy as np

# Sample data
data = np.random.normal(loc=0, scale=1, size=1000)

# Create distplot
fig = ff.create_distplot([data], group_labels=["Data"], show_hist=True, show_rug=True)
fig.update_layout(title="Distribution Plot with Plotly")
fig.show()
```
