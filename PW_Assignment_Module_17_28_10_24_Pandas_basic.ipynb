Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In Pandas, a DataFrame is a two-dimensional, size-mutable, and labeled data structure with columns of potentially different data types. It can be thought of as a table (like an Excel spreadsheet) with rows and columns, where each column represents a Pandas Series.

A Series, on the other hand, is a one-dimensional array-like structure that holds a single column of data with an associated index.

Key Differences:

Dimension: A DataFrame is two-dimensional (rows and columns), while a Series is one-dimensional.

Structure: A Series represents a single column of data, whereas a DataFrame is a collection of Series (multiple columns).

Data Types: A DataFrame can contain columns of different data types (e.g., integers, floats, strings), while a Series typically contains a single data type (although it can hold mixed types).

```python
# Example: DataFrame
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Claire"],
    "Age": [25, 30, 27],
    "Gender": ["Female", "Male", "Female"]
})
print(df)
```

```python
# Example: Series
age_series = pd.Series([25, 30, 27], name="Age")
print(age_series)
```

In this example,

The DataFrame df contains multiple columns: "Name," "Age," and "Gender."
The Series age_series holds only a single column of data, representing ages, with no additional column for names or gender.

Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

Answer:

Pandas provides a wide range of functions for data manipulation in a DataFrame. Here are some common ones:

1. head() and tail(): View the first or last few rows of a DataFrame.

Example: df.head(5) displays the first 5 rows, useful for quickly inspecting data.

2. info(): Provides a concise summary of the DataFrame, showing the data types, non-null counts, and memory usage.

Example: Use df.info() to get a quick summary of the dataset’s structure, especially useful for identifying missing values.

3. describe(): Calculates summary statistics for numerical columns, including count, mean, standard deviation, min, and max.

Example: Use df.describe() to understand the distribution and range of numerical data in the DataFrame.

4. ndrop(): Removes specified rows or columns.

Example: df.drop("Gender", axis=1) removes the "Gender" column from the DataFrame.

5. fillna(): Fills missing values with a specified value or method (like forward filling).

Example: df.fillna(0) replaces all missing values with 0, helpful when you need to avoid NaNs for calculations.

6. groupby(): Groups data by a specified column and performs aggregations on grouped data.

Example: df.groupby("Gender").mean() calculates the average for each gender category, useful for comparison purposes.

7. merge(): Combines two DataFrames based on a common key or index.

Example: Use pd.merge(df1, df2, on="ID") to join data from two DataFrames based on an "ID" column.

8. apply(): Applies a custom function to a DataFrame or Series.

Example: df["Age"].apply(lambda x: x + 1) adds 1 to each value in the "Age" column, useful for transformations.

```python
# Example:
# Sample DataFrame
data = pd.DataFrame({
    "Name": ["Alice", "Bob", "Claire", "David"],
    "Age": [25, 30, 27, 22],
    "Gender": ["Female", "Male", "Female", "Male"]
})

# Grouping by "Gender" and calculating the mean age
average_age_by_gender = data.groupby("Gender")["Age"].mean()
print(average_age_by_gender)
```

Q7. Create a DataFrame using multiple Series. Explain with an example.

Answer:


To create a DataFrame from multiple Series,  pass each Series as a column in a dictionary. Each Series will represent a column, and the DataFrame will use their indices as row labels if they align.

Example

Suppose we have three Series representing "Name," "Age," and "Gender" information. We will combine them to form a DataFrame.

```python
# Example
import pandas as pd

# Creating Series
name_series = pd.Series(["Alice", "Bob", "Claire"])
age_series = pd.Series([25, 30, 27])
gender_series = pd.Series(["Female", "Male", "Female"])

# Creating DataFrame from multiple Series
data_frame = pd.DataFrame({
    "Name": name_series,
    "Age": age_series,
    "Gender": gender_series
})

# Displaying the DataFrame
data_frame
```

Explanation:

Each Series represents a column:

1. name_series becomes the "Name" column.
2. age_series becomes the "Age" column.
3. gender_series becomes the "Gender" column.

Since each Series has the same length, they align perfectly by their indices to form rows in the DataFrame.

This method is useful when data is collected separately in Series but needs to be combined into a single structure for analysis or manipulation.
