Q5. A survey was conducted to examine the relationship between age and preference for a particular brand of soft drink. The survey results are shown below:

Age(Years)         Soft drink Preference
25                         Coke
42                         Pepsi
37                    Mountain dew
19                       Coke
31                      Pepsi
28                      Coke

Answer:

To analyze the relationship between age and soft drink preference based on the provided survey results, we can summarize the data and explore patterns. Since soft drink preference is categorical and age is continuous, we can examine how preferences differ by age group or create a contingency table.

Here's how we can approach this:

Summarize the Data: List the age and corresponding soft drink preference.
Categorize Age Groups: Optionally, we can categorize ages into groups (e.g., 18-24, 25-34, 35-44) to analyze trends.
Count Preferences by Age Group: Create a frequency table or count how many people in each age group prefer each drink.
Step 1: Summarize the Data
Age (Years)	Soft Drink Preference
25	Coke
42	Pepsi
37	Mountain Dew
19	Coke
31	Pepsi
28	Coke
Step 2: Count Preferences
Next, we can count how many participants preferred each drink:

Coke: 3 (ages 25, 19, and 28)
Pepsi: 2 (ages 42 and 31)
Mountain Dew: 1 (age 37)
Step 3: Frequency Table
Hereâ€™s a simple frequency table based on the preferences:

Soft Drink	Count
Coke	3
Pepsi	2
Mountain Dew	1
Age Group Analysis (Optional)
If we want to categorize ages into groups, we can use these ranges:

18-24
25-34
35-44
Age Group Counts:

18-24: 1 (Coke)
25-34: 3 (Coke, Pepsi, Coke)
35-44: 2 (Pepsi, Mountain Dew)
Conclusion
From this analysis, we see that:

Coke is the most preferred drink among the participants, especially among younger individuals (19-34 years).
Pepsi has moderate preference and is popular among the 25-34 age group.
Mountain Dew has the least preference overall.

```python
import pandas as pd

# Create a DataFrame from the survey data
data = {
    'Age': [25, 42, 37, 19, 31, 28],
    'Soft Drink Preference': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']
}

df = pd.DataFrame(data)

# Count preferences
preference_counts = df['Soft Drink Preference'].value_counts()

# Display counts of each soft drink preference
print(preference_counts)

# Optionally, categorize ages into groups
age_groups = pd.cut(df['Age'], bins=[18, 25, 35, 45], labels=['18-25', '26-35', '36-45'])
df['Age Group'] = age_groups

# Count preferences by age group
grouped_counts = df.groupby('Age Group')['Soft Drink Preference'].value_counts()

print(grouped_counts)
```

Interpretation:


Preference Counts:

The output will show how many participants prefer each drink.
Grouped Preferences: The second output will provide insights into soft drink preferences by age group, allowing you to analyze trends in preferences related to age.
