Q4. What is the role of support vectors in SVM Explain with example

Answer:

Support vectors play a crucial role in the functioning of a Support Vector Machine (SVM). They are the data points that lie closest to the decision boundary (or hyperplane) and are pivotal in defining that boundary. The SVM algorithm uses these support vectors to maximize the margin between the classes.

Role of Support Vectors:
Defining the Margin: Support vectors are the points that directly influence the position and orientation of the decision boundary. The margin is the region between the hyperplane and the closest data points from either class. An SVM aims to maximize this margin, ensuring that the model generalizes well to new data.
Robustness: The decision boundary is only affected by the support vectors and not by any other data points in the dataset. This makes SVMs robust to outliers, as only a few points (support vectors) determine the model's decision function.
Generalization: Maximizing the margin defined by the support vectors helps the model generalize better to unseen data, reducing the risk of overfitting.
Example:
Imagine a simple 2D classification problem where we need to separate two classes (e.g., circles and squares). A linear SVM will try to draw a straight line (hyperplane) that best separates these two classes.

Support vectors are the closest points from each class to this decision boundary. They are critical because if they were removed or moved slightly, the position of the boundary would change.
All other points, which are further away from the decision boundary, do not affect the boundary directly.
Illustration:
Suppose we have the following data points:

Circles at (1, 2), (2, 3), and (3, 3).
Squares at (6, 6), (7, 8), and (8, 9).
If the decision boundary (hyperplane) is defined such that it separates the circles from the squares, only the closest data points from each class will be the support vectors. For example:

A circle at (3, 3) and a square at (6, 6) might be the closest points to the boundary and thus would be the support vectors.
These points will determine the margin, and the SVM algorithm will adjust the hyperplane to maximize this margin.

Visualization Insight:
Imagine a rubber band stretched between two groups of points (representing different classes). The rubber band will touch only the points on the boundary, not those in the middle of each group. These boundary points are like support vectors in an SVM‚Äîthey define the separation between the groups.

Key Points:
Support vectors are not just any points; they are the most critical points for defining the margin.
If you remove non-support vector points, the hyperplane remains unchanged, but if you remove or change a support vector, the margin and the boundary can shift.
In summary, support vectors are essential for determining the optimal hyperplane and thus influence the SVM's decision-making and generalization capabilities.

2. Marginal Plane
The marginal planes are the boundaries that define the margin around the hyperplane. These are equidistant from the hyperplane and touch the closest data points, known as support vectors.

Example:

If the hyperplane separates circles and squares, the marginal planes will be parallel lines that are equidistant from the hyperplane and touch the nearest data points on both sides.

3. Hard Margin
A hard margin SVM is used when the data is linearly separable, meaning there are no misclassifications. The margin is maximized with no tolerance for errors.

Example:

In a dataset where all circles are clearly on one side and all squares on the other without overlap, a hard margin SVM would perfectly separate them without any data points within the margin.

4. Soft Margin
A soft margin SVM allows for some misclassification or data points to be within the margin, providing flexibility when the data is not perfectly separable. This helps to create a more generalizable model.

Example:

If some circles are mixed with squares, a soft margin SVM would allow certain points to fall inside the margin or even be misclassified, while still trying to maximize the margin overall.

Visual Summary:
Hyperplane: The central decision boundary separating the classes.
Marginal Planes: Boundaries defining the margin around the hyperplane.
Hard Margin: No points are allowed within the margin; data must be perfectly separable.
Soft Margin: Some points are allowed within the margin, enabling the SVM to handle overlapping or noisy data.
Real-World Example:
Imagine trying to classify images of cats and dogs. In a simple case (hard margin), all cat images are on one side, and all dog images are on the other, with a clear line between them. In more complex cases (soft margin), some cat images might appear similar to dog images, so a few points are allowed to be within the margin or misclassified for the model to find a better generalization.

These graphs and explanations illustrate how SVMs use hyperplanes and margins to classify data, with soft and hard margins providing flexibility based on the data's separability.

```python
from sklearn import datasets
import pandas as pd
import numpy as np

# Load the iris dataset
iris = datasets.load_iris()

# Create a Pandas DataFrame
iris = pd.DataFrame(
    data=np.c_[iris['data'], iris['target']],
    columns=iris['feature_names'] + ['target']
)

species = []

# Iterate through the 'target' column and assign species names
for i in range(len(iris['target'])):
    if iris['target'][i] == 0:
        species.append("setosa")
    elif iris['target'][i] == 1:
        species.append('versicolor')
    else:
        species.append('virginica')

# Add the 'species' column to the DataFrame
iris['species'] = species

# Display the first few rows of the DataFrame
iris.head()
```

```python
import matplotlib.pyplot as plt

setosa = iris[iris.species == "setosa"]
versicolor = iris[iris.species=='versicolor']
virginica = iris[iris.species=='virginica']

fig, ax = plt.subplots()
fig.set_size_inches(13, 7) # adjusting the length and width of plot

# lables and scatter points
ax.scatter(setosa['petal length (cm)'], setosa['petal width (cm)'], label="Setosa Petal", facecolor="blue")
ax.scatter(versicolor['petal length (cm)'], versicolor['petal width (cm)'], label="Versicolor", facecolor="green")
ax.scatter(virginica['petal length (cm)'], virginica['petal width (cm)'], label="Virginica", facecolor="red")

ax.set_xlabel("sepal length (cm)")
ax.set_ylabel("sepal width (cm)")
ax.grid()
ax.set_title("Iris petals")
ax.legend()
```

```python
from sklearn.model_selection import train_test_split

X = iris.drop(['sepal length (cm)', 'sepal width (cm)', 'target', 'species'], axis=1)
y = iris.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.55, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
# Input the kernel from the user

kernels = ['linear', 'rbf', 'poly']

for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    print("Accuracy using {}:".format(kernel), accuracy_score(pred, y_test))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np  # Make sure NumPy is imported

# Input the kernel from the user

def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out

kernels = ['linear', 'rbf', 'poly']

for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)

    pred = model.predict(X_test)

    print("Accuracy using {}:".format(kernel), accuracy_score(pred, y_test))

    fig, ax = plt.subplots()
    # title for the plots
    title = ('Decision surface of SVC ' + model.kernel)
    # Set-up grid for plotting.

    # Access data using DataFrame indexing
    X0 = X['petal length (cm)'].values  # Access 'petal length (cm)' column and convert to NumPy array
    X1 = X['petal width (cm)'].values   # Access 'petal width (cm)' column and convert to NumPy array

    xx, yy = make_meshgrid(X0, X1)

    plot_contours(ax, model, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.set_ylabel('y label here')
    ax.set_xlabel('x label here')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
    ax.legend()
    plt.show()
```
